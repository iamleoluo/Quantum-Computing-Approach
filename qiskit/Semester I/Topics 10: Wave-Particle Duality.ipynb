## Classical Physics Topic: Wave-Particle Duality

### Classical Physics Explanation

**Wave-particle duality** is a fundamental concept in quantum mechanics that describes how particles, such as photons and electrons, exhibit wave-like and particle-like properties. In classical physics, waves and particles are distinct entities—waves spread out and interfere, while particles have definite positions and paths. However, quantum systems can display both behaviors depending on the experiment performed.

**Key principles include:**

1. **Interference**: Particles like electrons or photons can produce interference patterns when passed through a double slit, a behavior traditionally associated with waves.

2. **Particle-like Behavior**: In specific experiments, these identical particles behave as if they have discrete, localized positions (as in the photoelectric effect).

3. **Superposition**: Particles can exist in a superposition of states, which is part of the wave-like behavior where they exist in all possible positions or states until observed.

4. **Measurement**: When a quantum system is measured, its wave-like superposition collapses into a definite state, corresponding to particle-like behavior.

## Quantum Computing Approach

In quantum computing, **qubits** can naturally model wave-particle duality. A qubit in superposition (created by applying a **Hadamard gate**) behaves like a wave, existing in a combination of both possible states (|0⟩ and |1⟩) simultaneously, analogous to a particle traveling through both slits in a double-slit experiment. The **measurement** of the qubit forces it to "collapse" into one of the two states, mimicking particle-like behavior when observed.

Using quantum gates, we can simulate both wave-like superposition and particle-like collapse, helping to visualize wave-particle duality in quantum systems.



## Qiskit Program to Simulate Wave-Particle Duality

In this Qiskit program, we will simulate **wave-particle duality** by applying quantum gates to qubits, demonstrating how qubits in superposition behave like waves and how measurement forces them to collapse into a specific state, representing particle-like behavior.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer, AerSimulator, StatevectorSimulator, QasmSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
import matplotlib.pyplot as plt
```

## Detailed Program Breakdown

### Step 1: Create the Quantum Circuit

- We create a quantum circuit with **1 qubit**, which will simulate the behavior of a particle (like a photon or electron) undergoing wave-particle duality. This qubit will be placed in a superposition, representing wave-like behavior, and measured to observe particle-like behavior.

```python
# Step 1: Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)
```

### Step 2: Initialize the Qubit

- The qubit is initialized in the |0⟩ state, representing the particle in a definite state (similar to a particle localized at one position). At this stage, the particle behaves like a classical particle with a well-defined state.

```python
# Step 2: Initialize the qubit in the |0> state (representing the particle at a specific position)
qc.initialize([1, 0], 0)
```

### Step 3: Apply the Hadamard Gate (H Gate)

- We apply a **Hadamard gate** to place the qubit in a superposition of |0⟩ and |1⟩. This represents the **wave-like behavior** of a quantum particle, similar to a photon or electron traveling through both slits simultaneously in a double-slit experiment. The qubit exists in a combination of both states, simulating the **interference pattern** produced by waves.

$$
|\psi⟩ = \frac{1}{\sqrt{2}}(|0⟩ + |1⟩)
$$

The Hadamard gate allows us to simulate the wave-like nature of quantum particles, where the particle exists in multiple states until it is measured.

```python
# Step 3: Apply a Hadamard gate to create a superposition (representing wave-like behavior)
# This simulates the particle existing in a superposition of states, like traveling through both slits simultaneously
qc.h(0)
```

### Step 4: Visualize the Bloch Sphere

- We use the **Bloch sphere** to visualize the qubit's state after applying the Hadamard gate. The Bloch sphere shows the qubit in a **superposition** of |0⟩ and |1⟩, representing the wave-like behavior of the particle. The qubit's state will appear as a point on the equator of the Bloch sphere, showing that it exists in a superposition of states.

```python
# Step 4: Visualize the qubit's state on the Bloch sphere to observe the wave-like superposition
initial_state= StatevectorSimulator().run(qc).result().get_statevector()

plot_bloch_multivector(initial_state)
```

### Step 5: Measurement

- We measure the qubit to **collapse the superposition** into one of the definite states, |0⟩ or |1⟩. This represents the **particle-like behavior** of quantum systems, where the particle has a specific position or state when observed. The measurement forces the wave-like superposition to collapse into one of the two possible states, just like observing a particle forces it to behave like a classical particle.

```python
# Step 5: Measure the qubit to collapse the superposition into a specific state (representing particle-like behavior)
qc.measure_all()
```

### Step 6: Simulate the Measurement and Plot the Results

- We simulate the quantum circuit using the **Qasm simulator**, which runs the circuit multiple times (1024 shots) and records the measurement outcomes. These outcomes represent the **particle-like behavior** of the system after the superposition collapses.
- The results will show how often the qubit collapses to the |0⟩ state and how often it collapses to the |1⟩ state, analogous to observing which slit the particle went through in a double-slit experiment.

```python
# Step 6: Simulate the measurement and plot the results (showing the particle-like outcome)
simulator = QasmSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

```

```python
# Print the measurement results and plot the histogram
print(f"Measurement results: {counts}")
plot_histogram(counts)
```

```python
# Draw the quantum circuit
qc.draw(output='mpl')
```

- The **measurement** represents the collapse of the wave-like behavior into a specific, particle-like state. The qubit can only collapse into one state, representing the observation of the particle’s position or path.
