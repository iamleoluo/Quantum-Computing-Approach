### Classical Physics Topic: Conservation of Energy

### Classical Physics Explanation

In classical physics, **energy conservation** is a fundamental principle that states that the total energy in a closed system remains constant over time. Energy can change forms—such as between **kinetic energy** (the energy of motion) and **potential energy** (stored energy)—but the total amount remains the same. This principle applies to various systems, from mechanical systems like pendulums to electrical circuits and thermodynamic processes.

For example:

- **Pendulum**: A pendulum exhibits periodic motion where energy continuously transforms between potential energy (at the highest points) and kinetic energy (at the lowest point).

- **Springs**: In systems involving springs (such as a mass-spring system), potential energy is stored when the spring is compressed or stretched, while kinetic energy is maximized when the mass moves the fastest.

In quantum mechanics, energy states are quantized, and particles can transition between discrete energy levels. Using quantum computing, we can simulate the transitions between energy states, which model the energy exchange between kinetic and potential forms, following the principle of energy conservation.

### Quantum Computing Approach

In quantum computing, qubits can be used to simulate the energy levels of a system, and quantum gates can model transitions between these energy levels. The evolution of the qubit's state under the influence of quantum gates mimics the **exchange between kinetic and potential energy** in classical systems. Applying quantum gates like the Hadamard and Phase gates allows us to simulate the oscillatory energy exchange between different forms while conserving the total energy.

### Qiskit Program to Simulate Conservation of Energy

In this Qiskit program, we will use qubits to simulate a system where energy oscillates between **kinetic** and **potential forms**, such as in a pendulum or mass-spring system. Quantum gates will represent the transitions between energy states, and we will observe how the qubit’s state evolves while conserving total energy.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator, StatevectorSimulator, QasmSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
import matplotlib.pyplot as plt
```

### Detailed Program Breakdown

#### Step 1: Create the Quantum Circuit

- We begin by creating a **quantum circuit** with 1 qubit. The qubit will represent the energy state of the system, and its state will oscillate between representing **kinetic energy** and **potential energy** as it evolves over time.

```python
qc = QuantumCircuit(1)
```

#### Step 2: Initialize the Qubit

- We initialize the qubit in the |0⟩ state, representing a system where all the energy is stored as **potential energy**. In a pendulum, this corresponds to the moment when the pendulum is at its highest point and momentarily at rest, with all of its energy in the form of potential energy.

```python
# Step 2: Initialize the qubit in the |0> state (representing all potential energy in a pendulum at its highest point)
qc.initialize([1, 0], 0)
plot_bloch_multivector(qc)
```

#### Step 3: Apply the Hadamard Gate (H Gate)

- A **Hadamard gate** is applied to place the qubit in a **superposition** of |0⟩ and |1⟩ states. This represents the system exchanging energy between **potential** and **kinetic** forms. The qubit is now in a superposition, analogous to the pendulum moving between its highest and lowest points, where energy continuously transitions between potential and kinetic.

$$
|\psi⟩ = \frac{1}{\sqrt{2}} (|0⟩ + |1⟩)
$$

This gate simulates the **oscillation** of the system, where potential energy is gradually converted into kinetic energy as the pendulum accelerates downward.

```python
# Step 3: Apply a Hadamard gate to put the qubit in superposition (representing the exchange between kinetic and potential energy)
# This simulates the system oscillating between kinetic and potential energy
qc.h(0)
plot_bloch_multivector(qc)
```

#### Step 4: Apply a Phase Gate (P Gate)

- A **Phase gate** is applied to simulate the **periodic motion** of the system. The phase shift represents the passage of time and how the system’s energy distribution evolves. The phase shift in classical systems like pendulums or springs corresponds to the continuous oscillation between kinetic and potential energy.

A phase shift of $π/4$ corresponds to part of the system’s oscillation, where energy has transitioned between forms.

```python
# Step 4: Apply a Phase gate to simulate the change in energy distribution over time
# The phase shift represents the periodic motion and the energy exchange between kinetic and potential forms
qc.p(np.pi/4, 0)
plot_bloch_multivector(qc)
```

#### Step 5: Apply the Pauli-X Gate (X Gate)

- The **Pauli-X gate** is applied to flip the qubit’s state, representing the moment when the system reaches maximum **kinetic energy**. In the case of a pendulum, this corresponds to the lowest point of the pendulum’s swing, where all the potential energy has been converted into kinetic energy.

```python
# Step 5: Apply a Pauli-X gate to flip the qubit’s state (simulating maximum kinetic energy at the lowest point of the motion)
qc.x(0)
plot_bloch_multivector(qc)
```

#### Step 6: Apply Another Phase Gate

- Another **Phase gate** is applied to simulate further **oscillatory motion**. This models the continuous exchange of energy between kinetic and potential forms as the system oscillates. The system returns to a state where energy is a mix of potential and kinetic.

```python
# Step 6: Apply another Phase gate to simulate further oscillatory motion (continuous energy exchange)
qc.p(np.pi/4, 0)
plot_bloch_multivector(qc)
```

#### Step 7: Visualize the Bloch Sphere

- To visualize the system’s state after applying these quantum gates, we use the **Bloch sphere** representation. The Bloch sphere shows the qubit’s state in 3D space, allowing us to observe how the system’s energy evolves between potential and kinetic forms.

```python
# Step 7: Visualize the qubit's state on the Bloch sphere to observe the energy exchange
simulator = StatevectorSimulator()
job = simulator.run(qc).result()
final_state = job.get_statevector()

plot_bloch_multivector(final_state)
```

#### Step 8: Measurement

- Finally, we measure the qubit to observe whether the system’s energy is in a **kinetic** or **potential state** at a particular point in time. The measurement collapses the qubit into either |0⟩ (potential energy) or |1⟩ (kinetic energy).

```python
# Step 8: Measure the qubit to observe the final energy state (kinetic or potential)
qc.measure_all()
```

#### Step 9: Simulate the Measurement and Plot the Results

- We simulate the quantum circuit using the **Qasm simulator**, which runs the circuit multiple times and records the measurement outcomes. These results represent the probability of the system being in either a potential or kinetic energy state at the measurement time.

```python
# Step 9: Simulate the measurement and plot the results
simulator = QasmSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()
```

- Finally, we plot the measurement results using a **histogram**, which shows the probability of the system being measured in each energy state.

```python
# Print the measurement results and plot the histogram
print(f"Measurement results: {counts}")
plot_histogram(counts)
```

```python
# Draw the quantum circuit
qc.draw(output='mpl')
```
