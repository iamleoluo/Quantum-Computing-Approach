### Classical Physics Explanation

In classical physics, the **conservation of angular momentum** states that if no external torque acts on a system, the system's total angular momentum remains constant. This principle applies to spinning objects (e.g., figure skaters pulling their arms in to spin faster) and orbital motion (e.g., planets orbiting the sun). Angular momentum is a vector quantity with both magnitude and direction.

In quantum mechanics, angular momentum is also quantized, and quantum particles such as electrons have intrinsic angular momentum known as **spin**. Quantum systems can exist in states with definite or indefinite angular momentum, and we can use qubits to represent these spin states and their transformations. Simulating transitions between different spin states using quantum gates allows us to model angular momentum conservation and demonstrate how quantum systems behave under transformations.

### Quantum Computing Approach

In quantum computing, qubits can be used to simulate the **spin states** of particles and their angular momentum. Quantum gates can simulate the evolution of these states, demonstrating the **conservation of angular momentum** under certain operations. Applying appropriate gates allows us to model the transitions between different spin states and observe how angular momentum is conserved in quantum systems.



### Qiskit Program to Simulate Conservation of Angular Momentum

In this Qiskit program, we will simulate the conservation of angular momentum using qubits. We’ll represent the spin state of a particle and apply quantum gates to simulate changes in the system’s angular momentum. Using the Bloch sphere visualization, we can observe how the angular momentum vector evolves.

```python
from qiskit import QuantumCircuit
from qiskit_aer import QasmSimulator
from qiskit_aer import StatevectorSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np
```

### Detailed Program Breakdown

#### Step 1: Create the Quantum Circuit

- We start by creating a quantum circuit with **1 qubit**. This qubit will represent the spin state of a particle and its angular momentum. In this simulation, we will transform the qubit to show how angular momentum changes but is conserved under certain operations.

```python
# Step 1: Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)
qc.draw(output='mpl')
```

#### Step 2: Initialize the Qubit

- We initialize the qubit in the |0⟩ state, representing the particle’s **lowest angular momentum state** (analogous to spin-down). This state corresponds to the qubit having the minimum angular momentum along a chosen axis (e.g., the z-axis).

```python
# Step 2: Initialize the qubit in the |0> state (representing the lowest angular momentum state, spin-down)
# This corresponds to the particle having the minimum angular momentum along a given axis
qc.initialize([1, 0], 0)
qc.draw(output='mpl')
```

#### Step 3: Apply the Hadamard Gate (H Gate)

- Next, we apply a **Hadamard gate** to the qubit. This gate creates a **superposition** of the |0⟩ (spin-down) and |1⟩ (spin-up) states, representing the particle's combination of spin-up and spin-down angular momentum states. In classical physics, this corresponds to a state where the angular momentum vector is uncertain.

$$
|\psi⟩ = \frac{1}{\sqrt{2}} (|0⟩ + |1⟩)
$$

```python
# Step 3: Apply a Hadamard gate to put the qubit in superposition (representing an uncertain angular momentum state)
# The superposition corresponds to the particle having a combination of spin-up and spin-down states
qc.h(0)
qc.draw(output='mpl')
```

#### Step 4: Apply the Pauli-Y Gate (Y Gate)

- We apply a **Pauli-Y gate** to rotate the qubit around the y-axis of the Bloch sphere. This gate simulates a change in the **direction of the angular momentum vector**. In classical terms, it’s like changing the direction of the particle's spin.

```python
# Step 4: Apply a Pauli-Y gate to rotate the qubit (simulating a change in the direction of the angular momentum vector)
qc.y(0)
qc.draw(output='mpl')
```

#### Step 5: Apply the Pauli-X Gate (X Gate)

- The **Pauli-X gate** flips the qubit’s state, corresponding to changing the spin from **spin-down** to **spin-up**. This gate simulates a reversal of the particle's angular momentum along the z-axis, analogous to a particle flipping its spin from -½ to +½ in classical physics.

```python
# Step 5: Apply a Pauli-X gate to simulate a spin-flip (this models a change in angular momentum state, e.g., spin-down to spin-up)
qc.x(0)
qc.draw(output='mpl')
```

#### Step 6: Apply Another Hadamard Gate

- After applying the Pauli-X gate, we apply another **Hadamard gate** to return the qubit to a **superposition** of |0⟩ and |1⟩ states. This represents the system returning to an uncertain angular momentum state, but the total angular momentum is conserved.

```python
# Step 6: Apply another Hadamard gate to return the qubit to a superposition state (demonstrating conservation of angular momentum)
qc.h(0)
qc.draw(output='mpl')
```

#### Step 7: Visualize the Bloch Sphere

- To visualize the qubit’s state, we use the **Bloch sphere** representation. The Bloch sphere shows the orientation of the qubit's angular momentum vector in 3D space. After applying the sequence of gates, we can observe how the angular momentum vector evolves on the sphere.

```python
# Step 7: Visualize the qubit's state on the Bloch sphere
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

```python
# Draw the quantum circuit
qc.draw(output='mpl')
```

#### Step 8: Simulate the Circuit

- We simulate the quantum circuit using the **statevector_simulator** from Qiskit Aer. This allows us to obtain the qubit's state at the end of the circuit and visualize it on the Bloch sphere.

```python
# Step 8: Simulate the circuit using the statevector_simulator
simulator = StatevectorSimulator()
job = simulator.run(qc).result()
final_state = job.get_statevector()
```

#### Step 9: Visualize the Final State

- Finally, we visualize the qubit’s state using the **plot_bloch_multivector** function, which shows the qubit's position on the Bloch sphere. This visualization represents the particle’s angular momentum state after all the transformations.

```python
# Step 9: Print the state and visualize the Bloch sphere
plot_bloch_multivector(final_state)
```

### Classical Physics Parallel

In classical physics, the **conservation of angular momentum** means that the total angular momentum of a closed system remains constant if no external torque acts on it. In this Qiskit simulation, we use quantum gates to represent changes in the angular momentum vector, but after applying various transformations, the total angular momentum remains conserved.

- The **qubit's state** represents the **spin state** of a particle, which has angular momentum.

- The **Hadamard gate** creates a superposition of spin states analogous to an uncertain angular momentum direction.

- The **Pauli-Y and Pauli-X gates** represent rotations and spin flips, simulating changes in the direction of the angular momentum vector.

- The **Bloch sphere** visualization shows how the qubit's angular momentum vector evolves under these operations.
