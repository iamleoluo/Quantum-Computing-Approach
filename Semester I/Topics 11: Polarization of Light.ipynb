### Classical Physics Topic: Polarization of Light

### Classical Physics Explanation

**Polarization** refers to the orientation of the oscillations of light waves. In unpolarized light, the oscillations of the electric field are in all directions perpendicular to the direction of propagation. When light is polarized, these oscillations are restricted to a specific plane.

There are different types of polarization:

1. **Linear Polarization**: The electric field oscillates in a single plane.

2. **Circular Polarization**: The electric field rotates in a circular motion as the wave propagates.

3. **Elliptical Polarization**: A more general form where the electric field traces an ellipse.

**Polarizing filters** can restrict light to a particular polarization, and interactions like **reflection** or **scattering** can cause light to become polarized. Polarization is fundamental in optics, 3D movie glasses, and various communication systems.

## Quantum Computing Approach

In quantum computing, qubits can simulate polarization states. A qubit in a **superposition** can represent the oscillating electric field of light in multiple directions. At the same time, **quantum gates** such as the **Hadamard gate**, **Pauli-X**, and **Pauli-Z** gates can change the polarization state (from unpolarized to linear or circular polarization).

Using different quantum gates, we can simulate how the polarization of light changes when passing through polarizers or interacting with materials, demonstrating the transition between unpolarized and polarized states.



## Qiskit Program to Simulate Polarization of Light

In this Qiskit program, we simulate the **polarization of light** using a single qubit. Initially, the qubit is unpolarized (in a superposition state), and we apply quantum gates to simulate different types of polarization, such as linear and circular polarization. We then measure the qubit to observe its polarization state.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer, AerSimulator, StatevectorSimulator, QasmSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
import matplotlib.pyplot as plt
```

### Detailed Program Breakdown

#### Step 1: Create the Quantum Circuit

- We create a quantum circuit with **1 qubit**. This qubit will represent the **polarization state** of light, with quantum gates simulating changes in polarization as light passes through polarizers or interacts with materials.

```python
# Step 1: Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)

```

#### Step 2: Initialize the Qubit in the Zero State

- The qubit is initialized in the |0⟩ state, representing the light in a **polarized state** along a specific axis. Initially, we treat this as the reference state, representing light aligned along a particular axis before applying any quantum operations.

```python
# Step 2: Initialize the qubit in the |0> state (representing unpolarized light)
qc.initialize([1, 0], 0)

```

#### Step 3: Apply the Hadamard Gate (H Gate)

- We apply a **Hadamard gate** to the qubit, placing it in a **superposition** of |0⟩ and |1⟩. This superposition represents **unpolarized light**, where the electric field oscillates in all directions perpendicular to the direction of propagation.

$$
|\psi\rangle = \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)
$$

This superposition simulates light before it interacts with a polarizing filter or before its polarization changes due to interaction with materials.

```python
# Step 3: Apply a Hadamard gate to put the qubit in superposition (representing unpolarized light)
# This represents the light oscillating in all directions (unpolarized)
qc.h(0)

```

#### Step 4: Apply a Pauli-X Gate (X Gate)

- The **Pauli-X gate** is applied to simulate **linear polarization**. This gate corresponds to rotating the qubit’s state along the X-axis, representing the alignment of the light’s electric field along a specific plane. In this case, the electric field oscillates along a single direction, analogous to light passing through a linear polarizer.

```python
# Step 4: Apply a Pauli-X gate to simulate linear polarization along the X axis
# This gate simulates the effect of a polarizer aligning the light along a particular plane
qc.x(0)

```

#### Step 5: Apply a Phase Gate (P Gate) to Simulate Circular Polarization

- A **Phase gate** is applied to introduce a phase shift of π/2 radians, simulating **circular polarization**. In circular polarization, the electric field rotates as the wave propagates. The phase shift in the qubit corresponds to this continuous rotation of the polarization angle, resulting in a circularly polarized light wave.

```python
# Step 5: Apply a Phase gate to simulate circular polarization
# This phase shift corresponds to the electric field rotating in a circular motion
qc.p(np.pi/2, 0)

```

#### Step 6: Visualize the Bloch Sphere

- We use the **Bloch sphere** to visualize the qubit’s state after applying the quantum gates. The Bloch sphere shows how the qubit’s polarization state evolves as the gates are applied. Linear polarization corresponds to the qubit being aligned along one axis, while circular polarization corresponds to a continuous rotation on the Bloch sphere.

```python
# Step 6: Visualize the qubit's state on the Bloch sphere to observe the polarization state
final_state = StatevectorSimulator().run(qc).result().get_statevector()

plot_bloch_multivector(final_state)

```

#### Step 7: Measurement

- We measure the qubit to observe the **final polarization state** of the light. The measurement collapses the qubit into either the |0⟩ or |1⟩ state, representing whether the light is polarized along one axis or another. The measurement outcome depends on the combination of quantum gates applied, showing whether the light is linearly or circularly polarized.

```python
# Step 7: Measure the qubit to observe the polarization state (linear or circular)
qc.measure_all()

```

#### Step 8: Simulate the Measurement and Plot the Results

- We simulate the quantum circuit using the **Qasm simulator**, which runs the circuit multiple times and records the measurement outcomes. These results represent the **polarization state** of the light after passing through the simulated polarizers and phase shifts.

```python
# Step 8: Simulate the measurement and plot the results

result = QasmSimulator().run(qc,shot =1024).result()
counts = result.get_counts(qc)

```

- Finally, we plot the measurement results using a histogram, which shows the probability of the qubit being measured in each polarization state.

```python
# Print the measurement results and plot the histogram
print(f"Measurement results: {counts}")
plot_histogram(counts)

```

```python
# Draw the quantum circuit
qc.draw(output='mpl')
```
