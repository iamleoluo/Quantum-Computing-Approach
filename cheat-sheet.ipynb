## import package

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer, AerSimulator, StatevectorSimulator, QasmSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import numpy as np
import matplotlib.pyplot as plt
```

## Quantum circuit and gate

```python
# Step 1: Create a quantum circuit with 1 qubit
qc = QuantumCircuit(1)
# Step 2: Initialize the qubit in the |0> state
qc.initialize([1, 0], 0)
# Step 3: Apply a Hadamard gate to simulate the superposition 
qc.h(0)
```

## State-Vector Simulator

```python
#Visualize the qubit's state on the Bloch sphere to observe the wave nature and interference effects
state = StatevectorSimulator().run(qc).result().get_statevector()

plot_bloch_multivector(state)
```

## Measurement

```python
# Measure the qubit to observe the probability distribution (representing the interference pattern)
qc.measure_all()
```

## QASM Simulator

```python
# Simulate the measurement and plot the results
result = QasmSimulator().run(qc,shot =1024).result()
counts = result.get_counts(qc)
```

```python
# Print the measurement results and plot the histogram (representing the interference pattern)
print(f"Measurement results: {counts}")
plot_histogram(counts)
```

## Draw circuit

```python
# Draw the quantum circuit
qc.draw(output='mpl')
```
